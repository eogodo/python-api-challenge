```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
from scipy.stats import linregress
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = ('http://api.openweathermap.org/data/2.5/weather?appid={0}&units=imperial&q=').format(weather_api_key)

print('Beginning Data Retrieval')
print('--------------------------------')

# set up lists to hold reponse info
cloud = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 0
record_set = 1

# Loop through the list of cities and perform a request for data on each
for city in cities:
    
    try:
        response = requests.get(url + city).json()
        
        cloud.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response['wind']['speed']) 
        
        record += 1 
        if record==51: 
            record = 1
            record_set += 1 
        
        print(f'Processing Record {record} of Set {record_set} | {city}')
              
    except KeyError: 
        print('City not found. Skipping...')
        
print('--------------------------------')
print('Data Retrieval Complete')      
print('--------------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
output_data = pd.DataFrame((zip(cities,cloud,country,date,humidity,lat,lng,max_temp,wind_speed)),columns=['City', 'Cloudiness', 'Country', 'Date', 'Humidity','Lat','Lng','Max Temp','Wind Speed'])
output_data.info()
```

```python
# Output File (CSV)
output_file = output_data.to_csv('output_data/cities.csv',index=False,header=True)
output_data = pd.read_csv('output_data/cities.csv')
output_data.head()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def linear_regress(x,y):

    slope, intercept, rvalue, pvalue, stderr = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x,y)
    plt.plot(x,regress_values,"r-")
    
    print(f'The rsquared is : {rvalue**2}')
    return(line_eq)
```

```python
# Create Northern and Southern Hemisphere DataFrames
northern = output_data.loc[output_data['Lat']>=0]
southern = output_data.loc[output_data['Lat']<0]
northern.head()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
line_eq=linear_regress(northern['Lat'],northern['Max Temp'])
plt.annotate(line_eq,(5,-20),fontsize=15,color="red")
plt.title('City Latitude vs Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.show()
print('Observation 2: This regression shows that there is a strong negative correlation of latitudes and maximum temperatures in the Northern Hemisphere. As expected, as latitude increases, maximum temperatures decrease.')
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
line_eq=linear_regress(southern['Lat'],southern['Max Temp'])
plt.annotate(line_eq,(-30,45),fontsize=15,color="red")
plt.title('City Latitude vs Max Temperature')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.show()
print('Observation 3:Cities in the Southern Hemisphere have a weaker correlation with latitudes than cities in the Northern Hemisphere. As expected, there is a positive correlation between the maximum temperature and the proximity to the equator.')
```
